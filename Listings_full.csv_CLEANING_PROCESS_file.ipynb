# Below is a first-pass cleaning of "listings_full.csv"

```python
import pandas as pd
```

```python
# import csv file
listings_full_df = pd.read_csv("./Resources/listings_full.csv")
pd.set_option('display.max_columns', None)
listings_full_df.head()
```

```python
listings_full_df.shape
```

# NOTE, this drops the columns from the "Airbnb Data Deep Dive"'s "Fields not required" tab 

# AND "id"

```python
listings_full_df = listings_full_df.drop(columns=["id", "listing_url", "scrape_id", "name", "summary", "space", "description", "experiences_offered", "neighborhood_overview", "notes", "transit", "access", "interaction", "house_rules", "thumbnail_url", "medium_url", "picture_url", "xl_picture_url", "host_id", "host_url", "host_name", "host_thumbnail_url", "host_picture_url", "host_neighbourhood", "host_verifications", "smart_location", "country_code", "country", "state", "is_location_exact", "calendar_updated", "requires_license", "license", "jurisdiction_names", "last_scraped", "host_location", "zipcode", "host_location"])
listings_full_df.head()
```

```python
listings_full_df.shape
```

# Determine NaN counts for each column using a generator function:

```python
[*listings_full_df.isnull().sum().sort_values().items()]
```

## Now, since "neighbourhood_group_cleansed", "weekly_price", "monthly_price", and "square feet" are all about 99% NaNs we will need to drop these columns as well.  In addition, we also decided to drop columns with 1,500 or more NaNs.  These included "host_response_rate",  "host_response_time", "host_about":

```python
listings_full_df = listings_full_df.drop(columns=["neighbourhood_group_cleansed","weekly_price", "monthly_price", "square_feet", "host_acceptance_rate", "neighbourhood", "host_response_rate", "host_response_time", "host_about"])
```

```python
[*listings_full_df.isnull().sum().sort_values().items()]
```

# Since we've dropped the columns so many NaNs, we arrive at this list:

```python
[*listings_full_df.dtypes.items()]
```

```python
listings_full_df.shape
```

```python
listings_full_df.nunique()
```

```python
# We'll drop "host_since", as we determined it is not practical to create meaningful buckets to reduce relevant bins
listings_full_df.drop(columns=["host_since"], inplace=True)
listings_full_df.shape
```

# Now, since we know we will add a parced version of the amenities column later, we'll go ahead and drop that column. 

```python
listings_full_df.drop(columns=['amenities'], inplace=True)
[*listings_full_df.dtypes.items()]
```

```python
listings_full_df.shape
```

# Create a list of the objects/categoricals

```python
list_of_categoricals = listings_full_df.dtypes[listings_full_df.dtypes == "object"].index.tolist()
list_of_categoricals_df = listings_full_df[list_of_categoricals]
[*list_of_categoricals_df.isnull().sum().sort_values().items()]
```

# Delete columns with NaNs from the categorical columns and save in new dataframe:

```python
listings_full_df2 = listings_full_df.drop(columns=['market', 'cleaning_fee', 'first_review', "last_review", 'security_deposit'])
listings_full_df2.nunique()
```

```python
listings_full_df2.shape
```

```python
list_of_categoricals_updated = listings_full_df2.dtypes[listings_full_df.dtypes == "object"].index.tolist()
list_of_categoricals_updated
```

```python
# Create a OneHotEncoder instance
from sklearn.preprocessing import StandardScaler,OneHotEncoder
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(listings_full_df2[list_of_categoricals_updated]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(list_of_categoricals_updated)
encode_df.head()
```

```python
encode_df.shape
```

```python
# Merge one-hot encoded features and drop the originals
listings_full_df2 = listings_full_df2.merge(encode_df,left_index=True, right_index=True)
listings_full_df2 = listings_full_df2.drop(columns=['host_is_superhost',
 'host_has_profile_pic',
 'host_identity_verified',
 'street',
 'neighbourhood_cleansed',
 'city',
 'property_type',
 'room_type',
 'bed_type',
 'price',
 'extra_people',
 'has_availability',
 'calendar_last_scraped',
 'instant_bookable',
 'is_business_travel_ready',
 'cancellation_policy',
 'require_guest_profile_picture',
 'require_guest_phone_verification'])

listings_full_df2.head()
```
